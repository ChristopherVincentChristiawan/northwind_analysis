```python
import pandas as pd

orderdetailsdf = pd.read_excel(r'Order Details.xlsx')
orderdetailsdf.head()
```

```python
orderdetailsdf['Unit Price'] =  orderdetailsdf['Unit Price'] - (orderdetailsdf['Discount'] * orderdetailsdf['Unit Price'])
```

```python
orderdetailsdf.head()
```

```python
ordersdf = pd.read_excel(r'Orders.xlsx')
ordersdf.head()
```

```python
new_df = ordersdf.set_index('Order ID').combine_first(orderdetailsdf.set_index('Order ID')).reset_index()
new_df.head()
```

```python
new_df = new_df[['Order ID','Product', 'Order Date', 'Unit Price', 'Quantity']]
new_df
```

```python
productsdf = pd.read_excel(r'Products.xlsx')
productsdf.head()
```

```python
new_df['Supplier'] = new_df.Product.map(productsdf.set_index('Product Name')['Supplier'].to_dict())
```

```python
new_df
```

```python
new_df['Order Date'] = pd.to_datetime(new_df['Order Date'], format='%Y')
```

```python
new_df
```

```python
new_df['Order Date'] = new_df['Order Date'].dt.strftime('%Y')
```

```python
new_df
```

```python
new_df['Unit Price'] = new_df['Unit Price'] * new_df['Quantity']
```

```python
new_df
```

```python
new_df = new_df[['Order Date', 'Unit Price', 'Supplier' ]]
new_df
```

```python
new_df = new_df.groupby(by=['Order Date', 'Supplier'], sort=False, as_index=False)['Unit Price'].sum()
new_df.head(30)
```

```python
new_df2 = new_df.groupby(by=['Order Date'], sort=False, as_index=False)['Unit Price'].max()
new_df2
```

```python
new_df2['Supplier'] = new_df2['Unit Price'].map(new_df.set_index('Unit Price')['Supplier'].to_dict())
new_df2
```
