```python
import pandas as pd

orderdetailsdf = pd.read_excel(r'Order Details.xlsx')
ordersdf = pd.read_excel(r'Orders.xlsx')

orderdetailsdf.head()
```

```python
ordersdf.head()
```

```python
orderdetailsdf['Unit Price'] =  orderdetailsdf['Unit Price'] - (orderdetailsdf['Discount'] * orderdetailsdf['Unit Price'])
orderdetailsdf['Unit Price'] = orderdetailsdf['Unit Price'] * orderdetailsdf['Quantity']
```

```python
orderdetailsdf.head()
```

```python
new_df = ordersdf.set_index('Order ID').combine_first(orderdetailsdf.set_index('Order ID')).reset_index()
```

```python
new_df.head()
```

```python
new_df = new_df[['Order ID', 'Ship Country', 'Unit Price', 'Quantity']]
new_df
```

```python
new_df.rename(columns={'Unit Price':'Total Price', 'Quantity':'Volume'},inplace=True)
```

```python
new_df
```

```python
new_df.groupby(['Ship Country'], sort=False, as_index=False)['Total Price'].sum().nlargest(10, 'Total Price')
```

```python
new_df.groupby(['Ship Country'], sort=False, as_index=False)['Volume'].sum()
```

### Terlihat bahwa USA merupakan wilayah dengan volume dan total penjualan paling tinggi. 
#### note: Volume disini merujuk ke jumlah item yang terjual dan total penjualan merupakan jumlah item dikalikan dengan harga per item dikurangi diskon.
