```python
import pandas as pd

orderdetailsdf = pd.read_excel(r'Order Details.xlsx')
orderdetailsdf.head()
```

```python
orderdetailsdf['Unit Price'] =  orderdetailsdf['Unit Price'] - (orderdetailsdf['Discount'] * orderdetailsdf['Unit Price'])
```

```python
orderdetailsdf.head()
```

```python
#inspeksi data ordersdf 
ordersdf = pd.read_excel(r'Orders.xlsx')
ordersdf.head()
```

```python
new_df = ordersdf.set_index('Order ID').combine_first(orderdetailsdf.set_index('Order ID')).reset_index()
new_df.head()
```

```python
new_df = new_df[['Order ID', 'Order Date', 'Product', 'Unit Price', 'Quantity']]
```

```python
new_df
```

```python
new_df['Order Date'] = new_df['Order Date'].dt.strftime('%Y-%m')
```

```python
new_df
```

```python
new_df['Unit Price'] = new_df['Unit Price'] * new_df['Quantity']
```

```python
new_df.head()
```

```python
new_df = new_df.groupby(by=['Product', 'Order Date'], sort=False, as_index=False)
new_df = new_df['Unit Price'].sum()
new_df.head()
```

```python
final_df = new_df
```

```python
final_df = final_df.groupby(by=['Order Date'], sort=False, as_index=False)['Unit Price'].max()
```

```python
final_df
```

```python
final_df['Product'] = final_df['Unit Price'].map(new_df.set_index('Unit Price')['Product'].to_dict())
```
